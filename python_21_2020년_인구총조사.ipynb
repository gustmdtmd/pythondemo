```python
# '시군구별'이 아닌 데이터만 뽑아온다.
df = df[df['시군구별']!='시군구별'].copy()
df
```

```python
# Serise로 리턴
print(type(df['variable'])) # Series
#print(df['variable'])
print(type(df['variable'].str)) # StringMethods
print(df['variable'].str.split('.')[1][0])
print(df['variable'].str.split('.')[1][1])
print(type(df['variable'].str.split('.'))) # Series
print(df['variable'].str.split('.'))

# DataFrame으로 리턴
print(type(df['variable'].str.split('.', expand=True)))
print(df['variable'].str.split('.', expand=True))
```

```python
# 컬럼 추가
df['연도'] = df['variable'].str.split('.', expand=True)[0]
df['월'] = df['variable'].str.split('.', expand=True)[1]
df['성별'] = df['variable'].str.split('.', expand=True)[2]
```

```python
df.head()
```

```python
df.tail()
```

```python
# 모든 데이터 타입이 object이다.
df.info()
```

```python
# 성별에 None이 포함되어 있다.
df['성별'].unique()
```

```python
# .nunique는 None이 아닌 unique 개수만 반환한다.
df['성별'].nunique()
```

```python
# .fill()은 전체를 변경해준다.
# .fillna()는 None값만 변경해준다.
# '성별'에서 'None'을 '전체'로 바꾸어 준다.
df['성별'] = df['성별'].fillna('전체')
```

```python
df.head()
```

```python
df.tail()
```

```python
df['성별'].unique()
```

```python
df['성별'] = df['성별'].replace('1', '남').replace('2', '여')
df['성별'].unique()
```

```python
df.tail()
```

```python
df.sample(10)
```

```python
# Series에만 사용할 수 있다.
# 빈도수를 계산한다.
df['성별'].value_counts()
```

```python
df = df.rename(columns = {'variable':'기간', 'value':'출생아수'})
df.head()
```

```python
df['출생아수'].unique()
```

```python
# '출생아수' 컬럼에서 '-'을 찾기 위한 작업
df.sample(10)
```

```python
df['출생아수'] = df['출생아수'].replace('-', np.nan)
```

```python
df.sample(10)
```

```python
# '출생아수'의 데이터 타입을 확인한다. 
df.info()
```

```python
# astype()은 컬럼의 타입 변경을 해준다.
df['출생아수'] = df['출생아수'].astype(float)
df.info()
```
